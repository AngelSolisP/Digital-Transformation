```python
import numpy as np
import matplotlib.pyplot as plt

# Define the confusion matrix for the Bomb Threat Detector:
# cm = [[TN, FP],
#       [FN, TP]]
cm = np.array([[18, 4],
               [ 3, 5]])

# Start a new figure for plotting
plt.figure()

# Display the matrix as an image (heatmap)
# 'interpolation="nearest"' ensures each cell is shown as a solid block
# 'cmap="Blues"' applies a blue color scale
plt.imshow(cm, interpolation='nearest', cmap='Blues')

# Add a colorbar to show the mapping from values to colors
plt.colorbar()

# Label the x-axis ticks at positions 0 and 1
# and name them according to predicted classes
plt.xticks([0, 1], ['Predicted Safe', 'Predicted Threat'])

# Label the y-axis ticks at positions 0 and 1
# and name them according to actual classes
plt.yticks([0, 1], ['Actual Safe', 'Actual Threat'])

# Annotate each cell with its numeric value
# Loop over row indices (i) and column indices (j)
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, cm[i, j], ha='center', va='center')

# Add a title to the plot
plt.title('Bomb Threat Detector Confusion Matrix')

# Label the axes for clarity
plt.xlabel('Predicted Label')
plt.ylabel('True Label')

# Render the plot to the screen
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the confusion matrix for the AI Assassin Drone:
# cm = [[TN, FP],
#       [FN, TP]]
cm = np.array([[7, 4],   # 7 True Negatives, 4 False Positives
               [3, 6]])  # 3 False Negatives, 6 True Positives

# Start a new figure
plt.figure()

# Display the confusion matrix as a heatmap
# interpolation='nearest' shows each value as a solid block
# cmap='Oranges' applies an orange color scale
plt.imshow(cm, interpolation='nearest', cmap='Oranges')

# Add a colorbar to map colors to numeric values
plt.colorbar()

# Label Xaxis ticks at positions 0 and 1
# 'Predicted Civilian' corresponds to column 0
# 'Predicted Terrorist' corresponds to column 1
plt.xticks([0, 1], ['Predicted Civilian', 'Predicted Terrorist'])

# Label Yaxis ticks at positions 0 and 1
# 'Actual Civilian' corresponds to row 0
# 'Actual Terrorist' corresponds to row 1
plt.yticks([0, 1], ['Actual Civilian', 'Actual Terrorist'])

# Annotate each cell with its count
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, cm[i, j], ha='center', va='center')

# Add title and axis labels
plt.title('AI Assassin Drone Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')

# Render the plot
plt.show()
```
